```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import median
```

```python
# Save file path to variable
population_income = "county_census_debt_data_byState.csv"
```

```python
population_income_pd = pd.read_csv(population_income)
population_income_pd.head()
```

```python
x = population_income_pd["Population"]
y = population_income_pd["PovertyCount"]
```

```python
plt.scatter(x,y)
plt.title("Population vs Income")
plt.xlabel("Population")
plt.ylabel("")


plt.savefig("Population_Income.png")
plt.show()
```
